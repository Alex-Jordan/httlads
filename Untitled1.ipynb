```python
from bs4 import BeautifulSoup
import requests

# url="https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

page = open(r"C:/Users/santoshernandezr/factbook2017/fields/2147.html", encoding='utf-8').read()
page[:200]

# Make a GET request to fetch the raw HTML content
# html_content = requests.get(page).text

# Parse the html content
soup = BeautifulSoup(page, "lxml")
# print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.text)
```

```python
for link in soup.find_all("a"):
    print("Inner Text: {}".format(link.text))
    print("Title: {}".format(link.get("title")))
    print("href: {}".format(link.get("href")))
```

```python
gdp_table = soup.find("table", attrs={"id": "fieldListing"})
gdp_table_data = gdp_table.tbody.find_all("tr")  # contains 2 rows
print(gdp_table_data)

# gdp = gdp_table_data.td.find_all("a")
# print(gdp)

# Get all the headings of Lists
headings = []
for td in gdp_table_data[0].find_all("td"):
    # remove any newlines and extra spaces from left and right
    headings.append(td.b.text.replace('\n', ' ').strip())

print(headings)
```
